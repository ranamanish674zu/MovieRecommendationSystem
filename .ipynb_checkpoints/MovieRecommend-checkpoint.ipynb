# Exploratory data analysis

```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
Dataset.groupby('title').mean()
#so here we can see that main column is rating which can perfactly used to recommend a movie 
#concept is that here we will testing which movie get highest average rating
```

```python
#average of movies rating 
Dataset.groupby('title').mean()['rating']
```

```python
#so finding that  which movie has highest average rating 
Dataset.groupby('title').mean()['rating'].sort_values(ascending=False)
```

```python
Dataset.groupby('title').mean()['rating'].sort_values(ascending=False).head()
```

```python
#now we are finding a movie how many times watched (timestamp)
#or we are finding which are those movies who watched/rated  highest time 
Dataset.groupby('title')['rating'].count().sort_values(ascending=False)
```

```python
#we are creating the dataframe  of average ratings
ratings=pd.DataFrame(Dataset.groupby('title').mean()['rating'])
ratings
print(ratings.head())
```

```python
#let's add no of rating in ratings dataset 
ratings["no of ratings"]=Dataset.groupby('title')['rating'].count()
ratings #dataframe with title ,average rating ,and no of time rated or watched movie
```

```python
ratings.sort_values(by='rating',ascending=False)
#we can see which movie has full out of full average rating , has watched less time ......so that type of movie we will not 
#suggest to a customer .....so we have to discard all that type of movies 
```

```python
#we can plot a histogram on num of ratings to see the distrbution of numbers of ratings
plt.figure(figsize=(10,6))
plt.hist(ratings["no of ratings"],bins=70)
plt.title("graph between watched and rated movie")
plt.ylabel("How many time rating appears")
plt.xlabel("How many time rating is given")
#here we can conclude-> 0-10 time rated movies is having more frequencey , 
# that's means which movie are rated more time(or more rated movie) whihc is has less frequency 
#from 0-100 times watched has more frequency but this are only 15,20% or less of the total watched movie,so we can say this are 
# no good movie (not all time )for recommend.
```

```python
plt.hist(ratings['rating'],bins=70)
plt.xlabel("rating ")
plt.ylabel("how many time rated")
#we can see 5.0 rating watch less time 
# we can conlude the most good rating is around 3.0 ,,,because most of the people rated around 3
```

```python
#let's make a joint plot
sns.jointplot(x='rating',y="no of ratings",data=ratings,alpha=0.5)
```

# Creating Movie Recommendation

```python
Dataset.head()
```

```python
movieMat=Dataset.pivot_table(index='user_id',columns='title',values='rating')
movieMat 
#showing a pertucular person has given rating to which which movie
```

```python
ratings.sort_values('no of ratings',ascending=False).head()
#highly watched movies 
```

```python
#let's finding the user wise any movies rating 
Starwars_UserRating=movieMat['Star Wars (1977)']
Starwars_UserRating
```

```python
Starwars_UserRating.head()
```

```python
#now we are finding the correlation of a perticular movie with other movies ........which will highly correlate with the 
# given movie we can recommend that movie
Similar_to_StarWars=movieMat.corrwith(Starwars_UserRating)
# showing how correlated Starwar movie wiht other movies
Similar_to_StarWars
#obousaly this is a series we can convert that into a pandas dataframe which will easy to perform operation with pandas 
```

```python
#here NaN is that person who has not rated both movie 
```

```python
corr_of_StarWar=pd.DataFrame(Similar_to_StarWars,columns=["correlation"])
corr_of_StarWar
```

```python
corr_of_StarWar.dropna(inplace=True)#droping the nan values
```

```python
corr_of_StarWar
```

```python
corr_of_StarWar.head()
```

```python
corr_of_StarWar.sort_values('correlation',ascending=False).head(10)
#highly correlated movies wiht StarWArs movie
```

```python
# but there is issue is that let's say 500 people watch starwars and 5 people watch hollow reed those 5 people given to both of 
# movies 5 rating than obousaly correlation will will go very high for that movie ...so we will do here so filterig like we can 
# put condition "plese consider  that movie which have no of watched movies and no of rated movie will be above 100".
```

```python
cor_Starwasrs= corr_of_StarWar.join(ratings['no of ratings'])
cor_Starwasrs.head()
    #joining corrofStarwar and no of ratings 
```

```python
#movies which will be recommended for the Starwars 
cor_Starwasrs[cor_Starwasrs['no of ratings']>100].sort_values('correlation',ascending=False)
```
